```python
import os
import json
import pandas as pd
import requests
import psycopg2
```

```python
#data extraction
#Aggregated insurance
# Path for aggregated insurance data
path1 = "C:/Users/ADMIN/Desktop/phonepe/pulse/data/aggregated/insurance/country/india/state"

# List the states available in the data directory
agg_insur_list = os.listdir(path1)

# Initialize a dictionary to store data
columns1 = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Insurance_type": [],
    "Insurance_count": [],
    "Insurance_amount": []
}
# Iterate over each state in the insurance data directory
for state in agg_insur_list:
    cur_states = os.path.join(path1, state)
    agg_year_list = os.listdir(cur_states)

    # Iterate over each year available for the current state
    for year in agg_year_list:
        cur_years = os.path.join(cur_states, year)
        agg_file_list = os.listdir(cur_years)

        # Iterate over each file (quarter) in the current year directory
        for file in agg_file_list:
            cur_files = os.path.join(cur_years, file)
            with open(cur_files, "r") as data_file:
                data = json.load(data_file)

                # Extract relevant data and append to the dictionary
                for transaction in data["data"]["transactionData"]:
                    columns1["Insurance_type"].append(transaction["name"])
                    columns1["Insurance_count"].append(transaction["paymentInstruments"][0]["count"])
                    columns1["Insurance_amount"].append(transaction["paymentInstruments"][0]["amount"])
                    columns1["States"].append(state)
                    columns1["Years"].append(year)
                    columns1["Quarter"].append(int(file.strip(".json")))

# Convert the dictionary to a DataFrame
aggre_insurance = pd.DataFrame(columns1)
# Clean and format the 'States' column
aggre_insurance["States"] = aggre_insurance["States"].str.replace("andaman-&-nicobar-islands", "Andaman & Nicobar")
aggre_insurance["States"] = aggre_insurance["States"].str.replace("-", " ")
aggre_insurance["States"] = aggre_insurance["States"].str.title()
aggre_insurance["States"] = aggre_insurance["States"].str.replace(
    "Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu"
)
```

```python
# PostgreSQL connection
mydb = psycopg2.connect(
    host="localhost",
    user="postgres",
    password="malathi03",
    database="phonepe_data",
    port="5432"
)
cursor = mydb.cursor()
 #Function to create and populate a table
def create_and_populate_table(create_query, insert_query, data):
    cursor.execute(create_query)
    mydb.commit()
    for index, row in data.iterrows():
        cursor.execute(insert_query, tuple(row))
    mydb.commit()
```

```python
# Aggregated Insurance Table
create_query1 = '''CREATE TABLE IF NOT EXISTS aggregated_insurance (
                    States VARCHAR(50),
                    Years INT,
                    Quarter INT,
                    Insurance_type VARCHAR(50),
                    Insurance_count BIGINT,
                    Insurance_amount BIGINT)'''
insert_query1 = '''INSERT INTO aggregated_insurance (States, Years, Quarter, Insurance_type, Insurance_count, Insurance_amount)
                    VALUES (%s, %s, %s, %s, %s, %s)'''
create_and_populate_table(create_query1, insert_query1, aggre_insurance)
```
